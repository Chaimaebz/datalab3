## 2. Maken van een matrix van de "vergelijkbaarheid" van de top 20 (meest verkochtte producten) gebaseerd op wanneer die zijn verkocht (welk maand) 

#### Hercontroleren van structuur van datasets nu

```python
# Let's understand the data structure first
print("=== CUSTOMERS DATA ===")
print(f"Customers shape: {df_customers.shape}")
print(f"Customers columns: {df_customers.columns.tolist()}")
print("\nSample customers data:")
display(df_customers.head())

print("\n=== SALES DATA ===")
print(f"Sales shape: {df_product_sales.shape}")
print(f"Sales columns: {df_product_sales.columns.tolist()}")
print("\nSample sales data:")
display(df_product_sales.head())

print("\n=== PRODUCTS DATA ===")
print(f"Products shape: {df_products.shape}")
print(f"Products columns: {df_products.columns.tolist()}")
print("\nSample products data:")
display(df_products.head())
```

### Het maken van sales matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
# First, create the sales matrix from the raw sales data
print("=== CREATING SALES MATRIX ===")

# Identify month columns in the sales data
month_columns = [col for col in df_product_sales.columns if 'Euro' in col or any(month in col for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])]
print(f"Month columns found: {month_columns}")

# Create sales matrix with SKU as index and months as columns
sales_matrix = df_product_sales.set_index('sku')[month_columns].fillna(0)
print(f"Sales matrix shape: {sales_matrix.shape}")

# Calculate cosine similarity between products based on sales patterns
from sklearn.metrics.pairwise import cosine_similarity
product_similarity = pd.DataFrame(
    cosine_similarity(sales_matrix),
    index=sales_matrix.index,
    columns=sales_matrix.index
)

print("Product similarity matrix created!")

# Now calculate total sales per product and get top 20
total_sales_per_product = sales_matrix.sum(axis=1).sort_values(ascending=False)
top_20_by_sales = total_sales_per_product.head(20)

print("Top 20 products by total sales:")
print(top_20_by_sales)

# Create heatmap for actual top 20 products by sales
top_20_skus = top_20_by_sales.index
subset_similarity = product_similarity.loc[top_20_skus, top_20_skus]

plt.figure(figsize=(15, 12))
sns.heatmap(subset_similarity, 
            annot=True, 
            cmap='RdYlBu_r', 
            center=0.5,
            square=True,
            fmt='.2f')
plt.title('Product Similarity when bought Heatmap (Top 20 by Sales Volume)', fontsize=16)
plt.show()
```

## STOP HIER VOOR NU ! 

```python
# Find and visualize top product pairs by similarity
def find_top_similar_pairs(similarity_matrix, top_n=20):
    # Get upper triangle to avoid duplicates
    upper_triangle = np.triu(similarity_matrix.values, k=1)
    
    # Find indices of top similarities
    similarity_df = []
    for i in range(len(similarity_matrix)):
        for j in range(i+1, len(similarity_matrix.columns)):
            similarity_df.append({
                'Product_1': similarity_matrix.index[i],
                'Product_2': similarity_matrix.columns[j],
                'Similarity': upper_triangle[i, j]
            })
    
    similarity_df = pd.DataFrame(similarity_df)
    return similarity_df.nlargest(top_n, 'Similarity')

# Get top 20 most similar product pairs
top_pairs = find_top_similar_pairs(product_similarity, top_n=20)
print("Top 20 Most Similar Product Pairs:")
display(top_pairs)

# Visualize top pairs
plt.figure(figsize=(12, 8))
bars = plt.barh(range(len(top_pairs)), top_pairs['Similarity'], 
                color='skyblue', edgecolor='navy', alpha=0.7)

# Add product pair labels
labels = [f"{row['Product_1']} & {row['Product_2']}" for _, row in top_pairs.iterrows()]
plt.yticks(range(len(top_pairs)), labels)
plt.xlabel('Cosine Similarity Score')
plt.title('Top 20 Most Similar Product Pairs', fontsize=14)
plt.grid(axis='x', alpha=0.3)

# Add value labels on bars
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2, 
             f'{top_pairs.iloc[i]["Similarity"]:.3f}', 
             va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
